```python
import tkinter as tk
from tkinter import *

field_text=""

def add_to_field(sth):
    global field_text
    field_text=field_text+str(sth)
    field.delete("1.0","end")
    field.insert("1.0",field_text)

def calculate():
    global field_text
    result=str(eval(field_text))
    field.delete("1.0","end")
    field.insert("1.0",result)

def clear():
    global field_text
    field_text=""
    field.delete("1.0","end")
    
window=tk.Tk()
window.title("Calculator")
window.geometry("310x290")
window.configure(bg='#F0F4F7')  
field=tk.Text(window,height=3,width=22,font=("Times New Roman",20),bg='#FFFFFF', fg='#000000')
field.grid(row=1,column=1,columnspan=4)
button_bg='#E0E0E0'  # Light gray
button_fg='#000000'  # Black

btn_1=tk.Button(window,text="1",command=lambda: add_to_field(1),width=5,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_1.grid(row=4,column=1)

btn_2=tk.Button(window,text="2",command=lambda: add_to_field(2),width=5,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_2.grid(row=4,column=2)

btn_3=tk.Button(window,text="3",command=lambda: add_to_field(3),width=5,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_3.grid(row=4,column=3)

btn_4=tk.Button(window,text="4",command=lambda: add_to_field(4),width=5,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_4.grid(row=3,column=1)

btn_5=tk.Button(window,text="5",command=lambda: add_to_field(5),width=5,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_5.grid(row=3,column=2)

btn_6=tk.Button(window,text="6",command=lambda: add_to_field(6),width=5,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_6.grid(row=3,column=3)

btn_7=tk.Button(window,text="7",command=lambda: add_to_field(7),width=5,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_7.grid(row=2,column=1)

btn_8=tk.Button(window,text="8",command=lambda: add_to_field(8),width=5,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_8.grid(row=2,column=2)

btn_9=tk.Button(window,text="9",command=lambda: add_to_field(9),width=5,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_9.grid(row=2,column=3)

btn_0=tk.Button(window,text="0",command=lambda: add_to_field(0),width=5,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_0.grid(row=5,column=1)

#operation buttons

btn_plus=tk.Button(window,text="+",command=lambda: add_to_field("+"),width=5,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_plus.grid(row=4,column=4)

btn_minus=tk.Button(window,text="-",command=lambda: add_to_field("-"),width=5,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_minus.grid(row=5,column=4)

btn_multiplication=tk.Button(window,text="*",command=lambda: add_to_field("*"),width=5,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_multiplication.grid(row=3,column=4)

btn_division=tk.Button(window,text="/",command=lambda: add_to_field("/"),width=5,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_division.grid(row=2,column=4)

btn_clear=tk.Button(window,text="clear",command=lambda:clear(),width=5,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_clear.grid(row=5,column=3)

btn_decimal=tk.Button(window,text=".",command=lambda:add_to_field(".") ,width=5,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_decimal.grid(row=5,column=2)

btn_open_bracket=tk.Button(window,text="(",command=lambda:add_to_field("(") ,width=5,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_open_bracket.grid(row=6,column=1)

btn_close_bracket=tk.Button(window,text=")",command=lambda:add_to_field(")") ,width=5,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_close_bracket.grid(row=6,column=2)

btn_equal=tk.Button(window,text="=",command=lambda:calculate() ,width=13,font=("Times New Roman",14),bg=button_bg, fg=button_fg)
btn_equal.grid(row=6,column=3,columnspan=2)

window.mainloop()
```
